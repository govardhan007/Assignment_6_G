##### Use the network (edge list) 'Wiki-Vote.txt' given in Canvas

```python
# NOTE: The data is given as an edgelist
# Meaning: each edge (node1,node2) in one line of the input file
input_data = '/content/sample_data/Wiki-Vote.txt'
```

```python
# Load the data as a directed graph (1 Point)
# Note-1: networkx read_edgelist will read the graph as undirected graph
# Note-2: you can convert this directed graph to an undirected graph wherever necessary in this assignment

graph = nx.read_edgelist(input_data,create_using= nx.DiGraph())

```

##### Print the following basic properties of the given graph

```python
# Print the number of nodes and edges in the graph here (2 Points)
tot_edges=graph.number_of_nodes()
tot_nodes=graph.number_of_edges()
print("Number of edges in the graph are",tot_edges)
print("Number of nodes in the graph are",tot_nodes)
```

```python
# Print the number of connected components here (1 Point)
# NOTE: There is a networkx package for this
ugraph = graph.to_undirected()
nx.algorithms.components.number_connected_components(ugraph)

j=nx.algorithms.components.number_connected_components(ugraph)
print("The number of connected components are",j)
```

```python
# Print the number of nodes in each component here(1 Point)
i=j-24
for nodess in list(nx.connected_components(ugraph)):
  i += 1
  print('The Number of nodes in the component '+ str(i) + '  are ' + str(len(nodess)))
```

```python
# Delete all nodes from components which have less than 10 nodes in the graph here (1 Point)
nodes_10 = min(nx.connected_components(ugraph), key=len)


for r_node in nodes_10:
  graph.remove_node(r_node)
print("The number of nodes present in the graph after deleting some nodes", graph.number_of_nodes())
print("The number of nodes in the undirected graph are",nx.number_connected_components(ugraph))
```

```python
# Plot the degree distribution of the graph with a line graph here (5 Points)
# NOTE: You have to plot both in-degree and out-degree distribution in the same plot
# REFERENCE: https://networkx.org/documentation/networkx-2.3/auto_examples/drawing/plot_degree_histogram.html
# You have to get both in_degree and out_degree
# You have to use line plot instead of a histogram

import collections
import matplotlib.pyplot as plt

deg = sorted([d for n, d in graph.in_degree()], reverse=True)  # degree sequence
deg_tot = collections.Counter(deg)
indeg, incnt = zip(*deg_tot.items())
fig, ax = plt.subplots()

plt.plot(indeg, incnt, color="r",  label="indegree")
dout = sorted([d for n, d in graph.out_degree()], reverse=True)  # degree sequence
deg_out = collections.Counter(dout)
outdeg, outcnt = zip(*deg_out.items())

plt.plot(outdeg, outcnt, color="y",  label="outdegree")
plt.legend(loc="upper right")

plt.title("Distribution of Indegree and outdegree")
plt.xlabel("Degree")
plt.ylabel("Count")
```
